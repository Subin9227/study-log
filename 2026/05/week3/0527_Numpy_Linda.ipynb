#차원 교재 실습

````python
vector = np.array([1, 2, 3])  # 원본 벡터
print("vector:", vector)
print("vector.ndim:", vector.ndim)
print("---------")

column_vector = vector[:, np.newaxis]  # 전체 원소를 세로 방향으로 확장
print("column_vector:\n", column_vector)
print("column_vector.ndim:", column_vector.ndim)
print("---------")

first_element_column = vector[0, np.newaxis]  # 첫 번째 원소만 세로 방향으로 확장

print("first_element_column:\n", first_element_column)
print("first_element_column.ndim:", first_element_column.ndim)
print("---------")

range_column_vector = vector[1:3, np.newaxis]  # 1(두번째 요소), 3(네 번째 요소 바로 전 요소) 범위를 세로 방향으로 확장
print("range_column_vector:\n", range_column_vector)
print("range_column_vector.ndim:", range_column_vector.ndim)
````

### 3번 퀘스트

<aside>
👉 주어진 2차원 NumPy 배열에 대해, 축(axis)을 기준으로 최댓값을 구하고 최종 배열의 차원을 출력하는 코드를 작성하세요.

```python
import numpy as np
array = np.array([[3, 7, 2], [8, 4, 6]])
```

</aside>

````python
array = np.array([[3, 7, 2], [8, 4, 6]])
arr = []

#행 axis의 최댓값
for i in range(len(array)):
  a = np.max(array[i])
  arr = np.append(arr, a)
print(f"{arr}는 {arr.ndim}차원이다")
#[참고] 행의 최댓값은 np.max(array, axis=1)로 할 수 있다


#열 axis의 최댓값
result = np.max(array, axis=0)
print(f"{result}는 {result.ndim}차원이다")
````

#유니버설 함수 교재 실습



🤖 numpy에서 연산과 유니버설함수 <br>
1. 기본 연산 : 내부적으로 유니버설 함수를 호출하여 브로드캐스팅과 요소별 연산을 수행
-> 여러 배열을 한 번에 연산하려면 중간에 임시 배열(메모리)이 생성되어 비효율적일 수 있다.

2. 유니버설 함수 :

*   지정인자 : 계산 결과를 다른 메모리에 새로 만들지 않고 기존 배열에 바로 덮어쓸 수 있다.
*   축 지정 : 특정 차원을 기준으로 한 번에 연산 가능



---

*   array 에 추가하는 방법은 append <br> np.append(추가하려는 배열, 추가하려는 요소)



#유니버설 함수 미니퀘스트

### 1번 퀘스트

<aside>
👉 배열 `[1, 2, 3, 4]`에 대해 NumPy의 유니버설 함수를 사용하여 모든 요소를 제곱한 결과를 출력하는 코드를 작성하세요.

</aside>

````python
array = np.array([1,2,3,4])
a = 0

for i in array:
  a += np.power(i,2)
  print(a)

print("total:", a)
````

### 2번 퀘스트

<aside>
👉 다음 두 배열의 요소별 합을 계산하고, 결과를 새로운 배열이 아닌 기존 배열에 저장하는 코드를 작성하세요.

```python
array1 = np.array([10, 20, 30])
array2 = np.array([1, 2, 3])
```

</aside>

````python
array1 = np.array([10, 20, 30])
array2 = np.array([1, 2, 3])

for i in range(len(array1)):
  a = np.add(array1[i], array2[i])
  print(a)
  array1 = np.append(array1, a)

print(f"array1에 요소별 합을 저장하면: {array1}")
````
