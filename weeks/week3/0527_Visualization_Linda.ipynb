````python
import re

text = "안녕하세요!!! 저는 AI 모델-입니다. 12345 데이터를   정리해 보겠습니다."

#한글만 추출
# ^ : 한글만 남기기 (^가 없으면 한글만 제거)
# ㄱㅡㅣ가-힣 : 할글 / ㄱㅡㅣ : 자모
kor = re.sub(r"[^ㄱㅡㅣ가-힣\s]", "", text)
print("---한글 제거:---\n", kor, "\n")

#반대로 영어만 추출 & 제거하려면
# ^a-zA-Z\s
#[참고] https://jimmy-ai.tistory.com/353

blank = ' '.join(kor.split())
print("===추가로 공백을 하나로 정리하면:===\n",blank)
````

### 2번 퀘스트

<aside>
👉

**세 개의 그룹(Group A, Group B, Group C)** 에 대해 **각각 다른 평균을 가지는 데이터**를 생성하고, 이를 이용해 **다중 박스 플롯을 그리는 코드**를 작성하세요.

````python
np.random.seed(23) #난수 고정

#rand 함수는 표준 정규분포를 따르기 때문에 평균으로 0을 가진다
# 🤖 np.random.randn()은 평균 (mu = 0), 표준편차 (sigma = 1)인 종 모양의 정규 분포에서 난수들을 무작위로 추출
data1 = np.random.randn(100)
data2 = data1 +2 #그래서 data2는 평균이 2
data3 = data1 -2 #그래서 data3는 평균이 -2

plt.boxplot([data1, data2, data3])
plt.show()
````

````python
plt.boxplot(data1,
            #이상값 강조
            flierprops=dict(marker='o', markerfacecolor='red', markersize=10, linestyle='none'),
            #스타일 커스터마이징
            patch_artist=True, boxprops=dict(facecolor='lightblue'), #박스 생삭 설정
            whiskerprops=dict(color='yellow') #수염 색상 설정
            )
plt.show()
````

### 3번 퀘스트

<aside>

`gridspec`을 사용하여 **불규칙한 레이아웃의 서브플롯을 생성하고**, 각각 **선 그래프, 산점도, 막대 그래프, 히스토그램**을 그리세요.

````python
import matplotlib.gridspec as gridspec

x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.random.randn(100)
categories = ['A', 'B', 'C', 'D', 'E']
values = [3,7,5,2,8]

#gridspec를 활용한 불규칙한 레이아웃의 서브플롯 생성
fig = plt.figure()
gs = gridspec.GridSpec(2,2,figure=fig)

ax1 = fig.add_subplot(gs[0, 0])
ax1.plot(x, y1)

ax2 = fig.add_subplot(gs[0,1])
ax2.scatter(x, y2)

ax3 = fig.add_subplot(gs[1,0])
ax3.bar(categories, values)

ax4 = fig.add_subplot(gs[1,1])
ax4.hist(y2, edgecolor='black')

plt.show()
````

### 3번 퀘스트

<aside>
👉

**1번 퀘스트에서 생성한 시계열 데이터에서 이상치를 탐지하고, 이상치만 강조하여 그래프에 표시하는 코드**를 작성하세요.

이상치는 사분위수 범위(IQR)를 이용해 판단합니다.

```python
# 시계열 데이터 생성
np.random.seed(42)
date_range = pd.date_range(start="2023-01-01", periods=100, freq="D")
values = np.cumsum(np.random.randn(100))
```

</aside>

````python
q1 = df['Value'].quantile(0.25)
q3 = df['Value'].quantile(0.75)
iqr = q3 - q1
k = 0.5

lower_bound = q1 - k * iqr
upper_bound = q3 + k * iqr

#이상치 여부 판별
df["Outlier"] = (df['Value'] < lower_bound) | (df['Value'] > upper_bound)

#이상치에서 True만 가져오기
outliers = df[df['Outlier']]


#이상치 시각화
sns.lineplot(x='Date', y='Value', data=df)
sns.scatterplot(x="Date", y='Value', data=outliers, color='red')
plt.show()
````

이상값은 **IQR(사분위 범위)를 사용하여 탐지**하세요.

```python
# 샘플 데이터 생성
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)  # 평균 50, 표준편차 10인 정규 분포 데이터 생성
```

```python
np.random.seed(42)
```

````python
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=5, size=30)

#검정통계량, p-value
t, p = stats.ttest_1samp(sample_data, 52)

print(f"t-통계량 : {t}")

if p < 0.05:
  print(f"유의수준 5%에서 50과 52 두 그릅의 평균 차이가 통계적으로 유의미하다.")
else :
  print(f"유의수준 5%에서 50과 52 두 그릅의 평균 차이가 통계적으로 유의미하지 않다.")
````

````python
# 관측된 데이터 (Observed)
observed = np.array([50, 60, 90])

# 기대값 (Expected)
expected = np.array([66, 66, 66]) * (observed.sum() / np.sum([66, 66, 66]))


#카이제곱 검정
chi2, p = stats.chisquare(observed, expected)

print(f"카이제곱 통계량: {chi2:}")
print(f"유의확률 : {p}")
````

````python
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
group3 = np.random.normal(loc=60, scale=10, size=30)

f, p = stats.f_oneway(group1, group2, group3)

print('f값이 크면 그룹과 그룹 사이의 평균 차이(집단 간 분산)가 압도적으로 크다는 뜻입니다.')
print(f"3개의 그룹의 평균을 비교하면 {f}\n")

print("p값은 내 데이터가 뻥(착시)일 확률로 0에서 1 사이 값을 가집니다.")
if p < 0.05:
  print(f"{p}는 0.05보다 작으므로, 통계적으로 유의미합니다.")
else:
  print(f"{p}는 0.05보다 크므로, 통계적으로 유의미하지 않습니다.")
````
